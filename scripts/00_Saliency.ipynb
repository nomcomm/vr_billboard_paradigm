```python
import pandas as pf
import os
import matplotlib.pyplot as plt
%matplotlib inline
from utils import pySaliencyMap
import cv2
```

```python
stimulus_dir = '../data/stimuli/'

images = [f for f in os.listdir(stimulus_dir) if f.endswith('.jpg')]
images.sort()
images.remove('all_images.jpg')
images
```

```python
saliency_values_by_image = []

for img_no in range(len(images)):
    file_path = stimulus_dir + images[img_no]
    print(file_path)
    img = cv2.imread(file_path)
    imgsize = img.shape
    img_width  = imgsize[1]
    img_height = imgsize[0]
    sm = pySaliencyMap.pySaliencyMap(img_width, img_height)

    saliency_map = sm.SMGetSM(img)
    binarized_map = sm.SMGetBinarizedSM(img)
    salient_region = sm.SMGetSalientRegion(img)

    plt.subplot(2,2,1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Input image')
    plt.subplot(2,2,2), plt.imshow(saliency_map, 'gray')
    plt.title('Saliency map')
    plt.subplot(2,2,3), plt.imshow(binarized_map)
    plt.title('Binarilized saliency map')
    plt.subplot(2,2,4), plt.imshow(cv2.cvtColor(salient_region, cv2.COLOR_BGR2RGB))
    plt.title('Salient region')

    plt.show()

    print(saliency_map.mean())
    saliency_values_by_image.append(saliency_map.mean())
```

```python
import pandas as pd

lst3 = []
percentile_list = pd.DataFrame({'Stimuli' : images,
                                'Saliency' : saliency_values_by_image,
                                'Recall' : [lst3] }, 
                                columns=['Stimuli','Saliency', 'lst1Tite'])

percentile_list
```
